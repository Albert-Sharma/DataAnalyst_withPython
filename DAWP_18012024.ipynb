```python
from matplotlib import pyplot as plt #alias 
x = [1,2,3,4,5,6,7,8,9]
y = [1,4,9,4,3,10,4,2,7]
plt.plot(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.show()
```

```python
plt.hist(x) # for descrete data
plt.show()
```

```python
b = [10,5,8,4,2]
plt.hist(b)
plt.show()
```

```python
#scattered plot
plt.scatter(x,y)
plt.show()
```

```python
plt.plot(x,y,'o') #aslo see how to do this with line style
```

```python
label = ['one','two','three','four']
sizes = [35,35,30,45]
plt.pie(sizes, labels=label, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart example')
plt.show()
```

```python
#area plot
q = [1,2,3,4,5,6,7,8,9]
w1,w2 =[4,20,4,2,5,1,4,9,9],[9,5,4,7,1,9,2,0,40]
plt.fill_between(q,w1,w2,color='skyblue',alpha=0.4,label= 'Area 1-2')
plt.plot(q,w1,label='line1',marker='o')
plt.plot(q,w2,label='line2',marker='o')
plt.xlabel('X-axis'),plt.ylabel('Y-axis'),plt.title('Area Chart example')
plt.legend(),plt.show()
```
