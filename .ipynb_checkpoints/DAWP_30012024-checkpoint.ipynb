```python
flight = sns.load_dataset('flights')
```

```python
flight
```

```python
flights = flight.pivot_table(index='month', columns='year',values= 'passengers')
```

```python
sns.heatmap(flights, cmap='Blues', linecolor='white')
```

```python
#Pair grid
## It gives the bunch of plots that is related to the data where we get all information together.
```

```python
iris_df2 = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris_df2, hue='species')
iris_g.map(plt.scatter) #color shows differnet type of species
```

```python
iris_g1 = sns.PairGrid(iris_df2, hue ='species')
iris_g1.map_offdiag(plt.hist)
```
