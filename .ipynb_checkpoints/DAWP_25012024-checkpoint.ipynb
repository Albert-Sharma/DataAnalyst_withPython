```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
crash_df = sns.load_dataset("car_crashes")
```

```python
crash_df
```

```python
sns.distplot(crash_df["not_distracted"])
```

```python
'''
Kernel desnsity estimation(KDE) is the line above 
What is KDE?
'''
```

```python
sns.distplot(crash_df["not_distracted"],kde=False)
```

```python
sns.distplot(crash_df["not_distracted"],kde=False,bins=20)
```

```python
#Join plot
# It will be used to comapre the two distribution plotss and a plits by scatter plot
sns.jointplot(x="speeding",y="alcohol", data = crash_df, kind = 'reg')
```

```python
sns.jointplot(x="speeding",y="alcohol", data = crash_df, kind = 'hex')
```

```python
#KDE plot
sns.kdeplot(crash_df['alcohol'])
```

```python
#pair plot,pair plot is going to give the relationship accross the data frames numerical values
sns.pairplot(crash_df, hue='alcohol')
```

```python
#categorical plot
# BAR PLOT
sns.barplot(x="sex",y="total_bill",data=trips_df)
```

```python
sns.barplot(x="sex",y="total_bill",data=trips_df, estimator = np.mean)
```

```python
sns.barplot(x="sex",y="total_bill",data=trips_df, estimator = np.median)
```

```python
sns.barplot(x="sex",y="total_bill",data=trips_df, estimator = np.var)
```

```python
sns.barplot(x="sex",y="total_bill",data=trips_df, estimator = np.cov)
```

```python
sns.barplot(x="sex",y="total_bill",data=trips_df, estimator = np.std)
```

```python
# we have tip data, want to change feature name tip to tip_data
trips_df.rename(columns={'tip': 'tip_data'},inplace = True)
```

```python
trips_df
```

```python
#count plot
sns.countplot(x="sex", data=trips_df)
```

```python
#box plot
#compare across the quartiles
sns.boxplot(x="day",y="total_bill", data= trips_df, hue= "sex")
```

```python
sns.boxplot(x="day",y="total_bill", data= trips_df, hue= "sex")
plt.legend(loc=0) # brings the legends in the left top 
```

```python
#violin plot
''' Combination of hte box plot and the kde plot'''
sns.violinplot(x="day", y="total_bill", data= trips_df, hue="sex")
plt.legend(loc=1) # here legends position is not changing why?
```

```python
sns.violinplot(x="day", y="total_bill", data= trips_df, hue="sex", split=True)
```

```python
#strip plot
sns.stripplot(x="day",y="total_bill",data = trips_df)
```

```python
sns.stripplot(x="day",y="total_bill",data = trips_df, jitter=True)
```

```python
sns.stripplot(x="day",y="total_bill",data = trips_df, jitter=True, hue="sex")
plt.legend(loc=0)
```

```python
sns.stripplot(x="day",y="total_bill",data = trips_df, jitter=True, hue="sex", dodge = True)
plt.legend(loc=0)
```

```python
#swarm plot
sns.swarmplot(x="day", y="total_bill", data= trips_df, hue="sex")
plt.legend(loc=0)
```
